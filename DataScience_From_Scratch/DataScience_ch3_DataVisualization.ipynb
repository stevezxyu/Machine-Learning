```python
'''
資料視覺化
data visualization
'''
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# 建立一個折線圖， x軸是年份， y軸是 GDP
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# 加個標題
plt.title("Normal GDP")

# 在 y 軸上加的標籤
plt.ylabel("Bilions of $")
plt.show()
```

```python
# 長條圖
from matplotlib import pyplot as plt

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars= [5, 11, 3, 8, 10]

# 長條圖的預設寬度為 0.8，所以我們在橫座標添加 0.1
# 長條圖將會對準中央的位置
xs = [i + 0.1 for i, _ in enumerate(movies)]

# 根據 x 座標 [xs] 與高度值 [num_oscars] 畫出長條圖
plt.bar(xs, num_oscars)

plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")

# 把電影名稱當作 x 軸的標籤，放在長條圖的中央位置
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

```python
# 直方圖
from matplotlib import pyplot as plt
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar([x - 4 for x in histogram.keys()],        # 每個長條圖都往左評移 4個單位
       histogram.values(),                        # 為每個長條圖設定正確的高度
       8)                                         # 將每個長條圖的寬度設定為 8個單位

plt.axis([-5, 105, 0, 5])                         # x 軸範圍設定為 -5到 105
                                                  # y 軸範圍設定為 0 到 5
plt.xticks([10 * i for i in range(11)])           # x 軸標籤設定為 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("# of students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

```python
from matplotlib import pyplot as plt

mentions = [500, 505]
years = [2013, 2014]

plt.bar([2012.6, 2013.6], mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

# 如果不這麼做，matplotlib 就會把 x 軸的標籤設定為 0, 1
# 然後在角落加個 +2.013e3
plt.ticklabel_format(useOffset = False)

# 如果 y 軸只顯示超出 500以上的部分，很容易造成誤導的效果
# plt.axis([2012.5, 2014.5, 499, 506])
# plt.title("Look at the 'Huge' Increase!")
# plt.show()

# 較為合理的座標軸
plt.axis([2012.5, 2014.5, 0, 550])
plt.title("Not So Huge Anymore")
plt.show()
```

```python
# 折線圖
# use plt.plot()
from matplotlib import pyplot as plt

variance     = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error  = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# 多做幾次呼叫 plt.plot
# 在同一個圖形中顯示多個資料序列
plt.plot(xs, variance,     'g-',  label='variance')       # 綠色實線
plt.plot(xs, bias_squared, 'r-.', label='bias^2')         # 紅色點虛線
plt.plot(xs, total_error,  'b:',  label='total error')    # 藍色點線

# 針對每個序列都指定了標籤
# 因此可以自由設定圖利說明
# loc = 9 表示「中間偏上」的位置
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

```python
# 散點圖
# 兩組成對數據資料的關係
from matplotlib import pyplot as plt

friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# 每個點的標籤
for label, friends_count, minutes_count in zip(labels, friends, minutes):
    xy = (friend_count, minute_count),     # 把標籤放到相應的點上
    xytext = (5, -5),                      # 稍微平移一下
    textcoords= ('offset points')

plt.title("Daily Minutes vs. Number of Friends")
```

```python
# 自動設定刻度，會具有誤導性的圖形
from matplotlib import pyplot as plt

test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grades")
plt.ylabel("test 2 grades")
plt.show()

# 使用 plt.axis("equal") 可以正確顯示出測試2 變動較大的情況
```
