```python
# 重複取樣 (Bootstrap)
# data = get_sample(num_points=n)

'''
利用重複取樣計算出每組樣本的中位數
    觀察樣本點分布的情況
'''
##############################################################
# 之前寫的函數

# 中位數 (median)
def median(v):
    '''
    找出落在 v 最中央位置的數值
    '''
    n = len(v)
    sorted_v = sorted(v)
    midpoint = n // 2
    
    if n % 2 == 1:
        # if it's odd, directly feedback to median
        return sorted_v[midpoint]
    else:
        # if it's even, feedback to two central location of mean
        lo = midpoint - 1
        hi = midpoint
        return (sorted_v[lo] + sorted_v[hi]) / 2
    
##############################################################
import random

def bootstrap_sample(data):
    '''
    根據原數據的數量，進行重複取樣，隨機取出相同數量的新樣本組
    '''
    return [random.choice(data) for _ in data]

def bootstrap_statistic(data, stats_fn, num_samples):
    '''
    對數據進行 num_samples 次重複取樣，並計算出相應統計函數 states_fn 的值
    '''
    return [stats_fn(bootstrap_sample(data))
           for _ in range(num_samples)]

# 101 個點，全都非常接近 100
close_to_100 = [99.5 + random.random() for _ in range(101)]

# 101 個點，其中 50 個很接近 0，另外 50個很接近 200
far_from_100 = ([99.5 + random.random()] + [random.random() for _ in range(50)] +
               [200 + random.random() for _ in range(50)])

bootstrap_statistic(close_to_100, median, 100)
bootstrap_statistic(far_from_100, median, 100)
```

```python
# t-distribution
# degree of freedom is n > k
# use "normal cumulative distribution function"

##############################################################
# 之前寫的函數

from matplotlib import pyplot as plt
import math

def normal_cdf(x, mu=0, sigma=1):
    return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2

##############################################################

def p_value(beta_hat_j, sigma_hat_j):
    if beta_hat_j > 0:
        # 如果係數值為正，必須把
        # 「看到更大值」的機率乘以兩倍
        return 2 * (1 - normal_cdf(beta_hat_j / sigma_hat_j))
    else:
        # 否則就是「看到最小值」的機率乘以兩倍
        return 2 * normal_cdf(beta_hat_j / sigma_hat_j)
    
print("p_value : %.8f" % p_value(30.63, 1.174))    # ~0 (常數項)
print("p_value : %.8f" % p_value(0.972, 0.079))    # ~0 (朋友數量)
print("p_value : %.8f" % p_value(-1.868, 0.131))   # ~0 (工作時數)
print("p_value : %.4f" % p_value(0.911, 0.990))    # 0.3575 (博士) 不顯著
```

```python
'''
正規(則)化 (Regularization)
    在誤差項裡添加一個懲罰項
    當 beta越大，懲罰項越大
'''
# 脊迴歸 (ridge regression)
# 在誤差項中添加一個正比於 beta_i 平方合的懲罰項

# alpha 是一個超參數「hyperparameter」，它控制是懲罰項的影響程度
# 有時它也用「lambda」來表示，但已被 python 定義
def ridge_penalty(beta, alpha):
    return alpha * dot(beta[1:], beta[1:])

def squared_error_ridge(x_i, y_i, beta, alpha):
    '''
    估計誤差的平方，再加上對 beta 的脊懲罰項 (ridge penalty)
    '''
    return error(x_i, y_i, beta) ** 2 + ridge_penalty(beta, alpha)

def ridge_penalty_gradient(beta, alpha):
    '''
    純粹只估算脊懲罰項(ridge penalty) 所對應的梯度值
    '''
    return [0] + [2 * alpha * beta_j for beta_j in beta[1:]]

def squared_error_ridge_gradient(x_i, y_i, beta, alpha):
    '''
    計算出第 i 平方誤差項所對應的梯度值 (包含脊懲罰項)
    '''
    return vector_add(squared_error_gradient(x_i, y_i, beta),
                     ridge_penalty_gradient(beta, alpha))

def estimate_beta_ridge(x, y, alpha):
    '''
    運用梯度遞減的做法，套入脊迴歸
    其中懲罰項的「超參數」設定為 alpha
    '''
    beta_initial = [random.random() for x_i in x[0]]
    return minimize_stochastic(partial(squared_error_ridge, alpha=alpha),
                             partial(squared_error_ridge_gradient,
                                    alpha=alpha),
                             x, y,
                             beta_iniitial,
                             0.001)
```
