```python
# CALLABLE可調式參數
import pandas as pd
import numpy as np
df1 = pd.DataFrame(np.random.randn(6, 4),
    index=list('abcdef'), columns=list('ABCD'))
```

```python
df1
```

```python
df1.loc[lambda df: df.A > 0,  :]
```

```python
df1.loc[ : , lambda df: ['A', 'B']]   # 列全部顯示 ; 顯示'A'欄和'B'欄
```

```python
df1.iloc[ : , lambda df: [0, 1]]    # 列全部顯示 ; 顯示第 0 欄和第 1欄
```
