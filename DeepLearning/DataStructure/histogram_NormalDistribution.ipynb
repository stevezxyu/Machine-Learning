```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.001)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# 建立一個 10000 normal 平均值為 2,0.5^2標準偏差
mu, sigma = 2, 0.5
v = np.random.normal(mu, sigma, 10000)
# 畫 500柱狀 histogram圖
plt.hist(v, bins=500, normed=1)
plt.show()
```
